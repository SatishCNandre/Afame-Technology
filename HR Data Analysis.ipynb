```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

The dataset comprises 1470 rows (ranging from 0 to 1469) and 35 columns, all of which are complete without any missing values, allowing us to move on to the next step.

#### Giving the columns new names.

```python
# Remove unnecessary columns
df = df.drop(columns=['EmployeeCount', 'Over18', 'StandardHours'])
```

```python
df
```

#### Rename Columns

```python
# Renaming columns for better clarity (example names)
df = df.rename(columns={
    'Attrition': 'EmployeeAttrition',
    'BusinessTravel': 'TravelFrequency',
    'DailyRate': 'DailyIncome',
    # Add more renaming if needed
})
```

#### Eliminate Redundant Entries

```python
# Remove duplicate rows
df = df.drop_duplicates()
```

#### Sanitize Specific Columns

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Trim whitespace from string columns
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

# Example of specific column sanitization
df['Department'] = df['Department'].str.title()  # Converting to title case
```

#### Eliminate NaN Values

```python
# Remove rows with any NaN values
df = df.dropna()
```

#### Check for Inconsistent Data Entries

```python
# Check unique values in categorical columns
categorical_columns = df.select_dtypes(include=['object']).columns
unique_values = {col: df[col].unique() for col in categorical_columns}
unique_values
```

#### Outlier Detection and Handling

```python
from scipy import stats

numerical_columns = ['DailyIncome', 'Age']
z_scores = stats.zscore(df[numerical_columns])
abs_z_scores = abs(z_scores)
filtered_entries = (abs_z_scores < 3).all(axis=1)
df_filtered = df[filtered_entries]
df_filtered
```

#### Encode Categorical Variables

```python
# Convert the categorical columns to 'category' dtype
df[categorical_columns] = df[categorical_columns].apply(lambda x: x.astype('category'))

# Encode categorical columns
df[categorical_columns] = df[categorical_columns].apply(lambda x: x.cat.codes)
```

```python
# Save the cleaned dataset
cleaned_file_path = 'HR_Data_Cleaned_Additional_Changes.csv'
df.to_csv(cleaned_file_path, index=False)
```

### Visualizations

```python
x = pd.read_csv('HR_Data_Cleaned_Additional_Changes.csv')
```

```python
x
```

#### Distribution of Age

```python
plt.figure(figsize=(10, 6))
sns.histplot(x['Age'], kde=True, color='green', bins=30)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

#### Attrition Rate by Department

```python
x['EmployeeAttrition'] = x['EmployeeAttrition'].astype(str)

plt.figure(figsize=(10, 6))
sns.countplot(x='Department', hue='EmployeeAttrition', data=x)
plt.title('Attrition Rate by Department')
plt.xlabel('Department')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

#### Correlation Heatmap

```python
plt.figure(figsize=(14, 12))
corr = x.corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

#### Bar Plot of Job Satisfaction

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='JobSatisfaction', data=df)
plt.title('Job Satisfaction Distribution')
plt.xlabel('Job Satisfaction Level')
plt.ylabel('Count')
plt.show()
```
